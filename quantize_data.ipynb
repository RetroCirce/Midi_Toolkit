```python
import numpy as np
import pretty_midi as pyd
from data_processor.midi_quantizer import MidiQuantizer
import os
```

```python
dataset_path = "../dataset/"
output_path = "../dataset/cleaned"
```

```python
# processed Multipianomide-Classic

dataset_name = "Multipianomide-Classic"

processor = MidiQuantizer(bpm=120, threshold=[60,200])


if not os.path.exists(os.path.join(output_path, dataset_name)):
    os.mkdir(os.path.join(output_path, dataset_name))

for folder in os.listdir(os.path.join(dataset_path,dataset_name)):
    f = os.path.join(dataset_path, dataset_name, folder)
    tf = os.path.join(output_path, dataset_name, folder)
    if not os.path.exists(tf):
        os.mkdir(tf)
    midi_files = os.listdir(f)
    midi_files = [d for d in midi_files if d.split(".")[0] + "_format0.mid" not in midi_files]
    for midi in midi_files:
        processor.process(os.path.join(f, midi), os.path.join(tf, midi))


    
```
